```python
import pandas as pd
passaporti = pd.read_csv("passaporti_completi.csv",sep=";")
```

```python
passaporti.columns
```

```python
places = passaporti.provenienza.unique()
```

```python
places
```

```python
migrations = ['africa','america','asia','austria','austra_ungheria','bosnia_erzegovina',
              'europa','esteri_europa','francia','germania','italia',
                'svizzera','turchia','ungheria']
```

```python
links = []
for p in places:
    if (pd.isna(p) == False):
        for m in migrations:
            link = {}
            link['source'] = p
            if (pd.isna(m) == False):
                link['target'] = m
                tot_m = passaporti[passaporti.provenienza == p][m].sum()
                link['value'] = tot_m
                links.append(link)
```
